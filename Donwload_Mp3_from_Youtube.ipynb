```python
import yt_dlp
from moviepy.editor import VideoFileClip
import os

def download_video(youtube_url, output_path="downloaded_video.mp4"):
    ydl_opts = {
        'format': 'mp4',
        'outtmpl': output_path,
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        print(f"Baixando vídeo de: {youtube_url}")
        ydl.download([youtube_url])
        print(f"Download concluído: {output_path}")

def extract_audio_segments(video_path, segments, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    video_clip = VideoFileClip(video_path)
    for i, (start_time, end_time) in enumerate(segments):
        print(f"Cortando segmento {i+1}: {start_time} a {end_time}")
        audio_clip = video_clip.subclip(start_time, end_time).audio
        output_audio_path = os.path.join(output_folder, f"segment_{i+1}.mp3")
        audio_clip.write_audiofile(output_audio_path)
        print(f"Áudio salvo como MP3: {output_audio_path}")
    video_clip.close()

def main():
    youtube_url = input("Insira a URL do vídeo do YouTube: ")
    output_video_path = "sample_data/downloaded_video.mp4"
    output_folder = "sample_data/audio_segments"

    download_video(youtube_url, output_video_path)

    segments = []
    while True:
        try:
            start_time = float(input("Insira o tempo de início (em segundos, ou '0' para terminar): "))
            if start_time == 0:
                break
            end_time = float(input("Insira o tempo de término (em segundos): "))
            segments.append((start_time, end_time))
        except ValueError:
            print("Por favor, insira tempos válidos em segundos.")

    extract_audio_segments(output_video_path, segments, output_folder)

    # Limpeza
    if os.path.exists(output_video_path):
        os.remove(output_video_path)
        print(f"Arquivo temporário removido: {output_video_path}")

if __name__ == "__main__":
    main()
```

```python
import os
import zipfile

# Caminho para o diretório que você deseja zipar
dir_to_zip = '/content/sample_data/audio_segments'
# Nome do arquivo ZIP de saída [deve ser diferente da pasta que está sendo zipada]
zip_filename = '/content/audio_segments.zip'

# Função para zipar um diretório
def zip_directory(directory_path, zip_filename):
    # Cria o arquivo ZIP
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # Caminha pelo diretório
        for root, dirs, files in os.walk(directory_path):
            for file in files:
                # Cria o caminho completo do arquivo
                file_path = os.path.join(root, file)
                # Adiciona o arquivo ao ZIP
                # O parâmetro arcname é usado para remover a parte do caminho que não é necessária
                zipf.write(file_path, arcname=os.path.relpath(file_path, directory_path))
    print(f'Arquivo ZIP criado: {zip_filename}')

# Chama a função
zip_directory(dir_to_zip, zip_filename)
```

```python
#Renomear arquivos da pasta sample_datas
import os

def rename_files(directory, new_names):
    """
    Renomeia arquivos no diretório especificado com os novos nomes fornecidos.

    Parameters:
    directory (str): O caminho para o diretório contendo os arquivos.
    new_names (list of str): Lista com os novos nomes dos arquivos.
    """
    # Verifica se o diretório existe
    if not os.path.exists(directory):
        print(f"O diretório '{directory}' não existe.")
        return

    # Obtém a lista de arquivos no diretório
    files = os.listdir(directory)

    # Conta o número de arquivos no diretório
    num_files = len(files)
    print(f"Total de arquivos encontrados no diretório: {num_files}")

    # Verifica se o número de novos nomes corresponde ao número de arquivos
    if len(new_names) != num_files:
        print("O número de novos nomes não corresponde ao número de arquivos.")
        print(f"Esperado: {num_files} novos nomes.")
        print(f"Fornecido: {len(new_names)} novos nomes.")
        return

    # Renomeia os arquivos
    for old_name, new_name in zip(files, new_names):
        old_file_path = os.path.join(directory, old_name)
        new_file_path = os.path.join(directory, new_name)

        # Renomeia o arquivo
        os.rename(old_file_path, new_file_path)
        print(f"Renomeado: '{old_name}' -> '{new_name}'")

# Exemplo de uso:
directory = '/content/sample_data/audio_segments'
new_names = ['new_name1.ext', 'new_name2.ext', 'new_name3.ext']  # Liste os novos nomes aqui

rename_files(directory, new_names)

```
